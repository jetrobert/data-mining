```python
import numpy as np
import tensorflow as tf
from tensorflow.contrib.learn.python.learn.estimators import kmeans as kmeans_ops
from tensorflow.contrib.learn import KMeansClustering as KMeans
from tensorflow.contrib.learn import RunConfig
```

# random centers generated func

```python
def make_random_centers(num_centers, num_dims):
    return np.round(np.random.rand(num_centers, num_dims).astype(np.float32) * 500)
```

# random points generated func

```python
def make_random_points(centers, num_points, max_offset=20):
    num_centers, num_dims = centers.shape
    assignments = np.random.choice(num_centers, num_points)
    offsets = np.round(np.random.randn(num_points,
                        num_dims).astype(np.float32) * max_offset)
    return (centers[assignments] + offsets, assignments, np.add.reduce(offsets * offsets, 1))
```

# random points

```python
num_centers = 6
num_dims = 2
num_points = 10000
true_centers = make_random_centers(num_centers, num_dims)
points, _, scores = make_random_points(true_centers, num_points)
```

# kmeans clustering

```python
kmeans = KMeans(num_centers, use_mini_batch=False, 
                config=RunConfig(tf_random_seed=14), random_seed=12)

kmeans.fit(x=points, steps=10, batch_size=8)
clusters = kmeans.clusters()

kmeans.predict(points, batch_size=128)
kmeans.score(points, batch_size=128)
kmeans.transform(points, batch_size=128)
```
